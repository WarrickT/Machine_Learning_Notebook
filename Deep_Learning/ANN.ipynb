## How do we train a Neuron

Given the following:  

$$
\text{input } \mathbf{x} \in \mathbb{R}^d, \text{ predicted output } y \in \mathbb{R}, \text{ ground truth } t \in \mathbb{R}, \text{ Neuron } M(\mathbf{x}; \mathbf{w})
$$

Follow this procedure. 

1. Make a $\textcolor{blue}{prediction}$ for some input data $\mathbf{x}$, with a known correct output $t$
   
   $$
   y = M(\mathbf{x}; \mathbf{w})
   $$

2. Compare the correct output with our predicted output to compute \textcolor{blue}{loss}
   
   $$
   \mathcal{L}(y, t)
   $$

3. $\textcolor{blue}{Adjust the weights/bias terms}$ to make the prediction closer to the ground truth, i.e., minimize loss $\mathcal{L}$

4. $\textcolor{blue}{Repeat}$ until we have an acceptable loss $\mathcal{L}$ / error $E$
