# Artificial Neurons: The basis of DL

The whole point of Deep Learning is to model the human brain. 

## Artificial Neuron Definition

$x_i$: Input (such as image pixel)

$w_i$: Weight for input $x$ that we learn 
for this particular input

$b$: Bias; a weight we learn with no input

$f$: **Activation Function** that determines how our output changes

$y$: **Output**



The goal is to express the output as a **weighted sum** of inputs and bias. 

## Activation Functions

We usually don't use linear activations, because multiple **linear layers** will just collapse into a single linear layer.

## The most common activation functions

### Sigmoid: Hyperbolic Tangent 

$\text{Hyperbolic Tangent: } f(x) = tanh(x)$

This clips values to -1 and 1. 

Signage is retained!



### Sigmoid: Logistic Function

$\text{Logistic Function: } f(x) = f(x) = \frac{1}{1 + e^{-x}}$

This clips values to 0 and 1. 

We went over this in Logistic Regression. 

Any **positive** value maps to [0.5, 1]

Any **negative** value maps to [0, 0.5)



### ReLU and friends 

Also called Rectified Linear Unit Activation Functions. 

$\mathrm{ReLU}(x) = (x)^+ = \max(0, x)$
