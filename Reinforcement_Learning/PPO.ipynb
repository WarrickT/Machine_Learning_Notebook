## Recall the following 

**Policy**: Mapping from action space to state space

A policy is essentially instructions for the RL agent, given the *state of the environment* it is currently in.

## What is PPO?



Recall that evaluating an agent means evaluating the policy function to find out how well the agent performs. 

When an agent is learning and doesn't know which actions give the best results, it does so by calculating the *policy gradients*.

## Recall the Objective Function

$$
J({\pi}_{\theta}) = {E}_{{\tau} ~ {\pi}_{\theta}}[R(\tau)]
$$

PPO aims to maximize the expected sumulative reward obtained from interacting with the environment. This is typically done by **maximizing a surrogate objective function** that approximates the policy improvement

## Clipped Surrogate Objective

This is the key feature of the PPO to prevent large policy updates. 

Clip the ratio between the probability of actions under the new policy and the old policy, PPO ensures that the policy update stays within a safe range. 

## Multiple Epochs and Mini-Batch Updates: 

PPO typically involves multiple epochs and mini batch updates to collect trajectories.
