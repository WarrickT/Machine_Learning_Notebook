## Basic Structure of Data

RL learns to act. It assumes input data to be *interdependent tuples*, aka order sequences of data. 

They are organized as state-action-reward tuples 

## Dynamics of Agent Environment 

We can define the dynamics of the environment with this mathematical expression 

$$
Prob(S_{t+1}=s', R_{t+1}=r|S_t=s, A_t = a)
$$

In simpler words and also a shorthand notation 

$$
p(s', r|s, a)
$$: 

"The probability of next state and reward being some values given some state and action to be taken"



## State: $S_t$

## Action: $A_t$

This is the action that the agent took. 

## Reward Signal: $r$

This is the reward for the action taken. 

For the dog, this would be maybe one treats, or two treats, or maybe even the whole bag. 

## Policy: $\pi (a | s)$

Probability the agent will *take action* when in a particular state. 

The agent will take action by sampling randomly from this distribution.

If $\pi$ has NO RANDOMNESS, only one action will be taken. This makes it deterministic.

## Return: $G_t$

$$
G_t = \sum_{k=t+1}^{T}\gamma^{k-t-1}R^{k} = R_{t+1} + {\gamma}R_{t+2} + {\gamma}^2R_{t+3} + ... + {\gamma}^{T-t-1}R_{T}

$$

**Dang** that's a lot of varaibles 

**Terminal State $T$**: This marks the end of an episode

## The goal is to maximize G_t

This may seem obvious, but our goal is to obtain the maximum of the *expected value* of this return function. 

$$

\max_{\pi} \mathbb{E}_{\pi}[G_t]

$$

## Random Detour to probability because I haven't taken a stats course yet :(

This is the **average outcome** of an event if it were to be repeated many times. 

It is calculated by multiple each possible outcome by its probability and summing these products . 

It is a *weighted average* of all possible outcomes, with the weights being the probabilities of those outcomes.

$$
\mathbb{E}(x) = x_1p_1 + x_2p_2 + ... + x_np_n 

$$

Mapping from *action* space to *state* space

## Trajectory

This is an entire sequence of $S_t, A_t, S_{t+1}, R_{t+1}$

## The State Value Functions

$V_{\pi}(s) = \mathbb{E}[G_t | S_t = s]$

In plain English, this is the expectation of the Reward given you're in the state $S_t$

## The action value function

$$

q_{\pi} = \mathbb{E}_{\pi}[G_t|S_t = s, A_t =a]

$$

## Now why all these functions? 


The whole point is to find the **optimal policy** such that: 

$\pi_{*}$ such that

$V_{\pi_*}(s)$  $V_{\pi}(s)$ for all $s$ and for any $\pi$

$q_{\pi_*}(s)$  $q_{\pi}(s)$ for all $s, a$ and for any $\pi$




## Some more useful defniitions: These are fancy ways to say some obvious things. 

**Tabular**: An environment is defined as **tabular** if all states and actions can be LISTED.

**Complete Knowledge**: This means we have complete access to environment dynamics. This is overaly ideal and we don't have access to this all the time. For example, Snake. 

**Policy**: Mapping from action space to state space

A policy is essentially instructions for the RL agent, given the *state of the environment* it is currently in.
